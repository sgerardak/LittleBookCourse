```python
import pandas as pd
from IPython.display import display
```

```python
stock_prices=pd.read_csv("stock_prices.csv")
stock_prices.head(25)
```

```python
stock_prices["yield_rate"]=stock_prices["close_one_year_later"]/stock_prices["close"]-1
stock_prices.head(25)
```

```python
# Extract the year from the 'date' column
stock_prices['year'] = pd.to_datetime(stock_prices['adjusted_date']).dt.year

# Calculate the average yield rate for each year
average_yield_per_year = stock_prices.groupby('year')['yield_rate'].mean().reset_index()

```

```python
display(average_yield_per_year)
```

```python
average_yield_per_year.to_csv("yearly_yield.csv")
```
